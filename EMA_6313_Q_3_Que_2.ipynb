```python
#Q-2 critical resolved shear stress
#UFID = 8 0 7 7 3 6 5 3 = D1 D2 D3 D4 D5 D6 D7 D8
#working code
import numpy as np
import math

#defining parameters
a = 0.1
b = 0.2
c = 0.3
alpha = 90
beta = 90
gamma = 90

#defining generalized metric tensor

y = np.array([[a**2, a*b*math.cos(math.radians(gamma)), a*c*math.cos(math.radians(beta))],
                [b*a*math.cos(math.radians(gamma)), b**2, b*c*math.cos(math.radians(alpha))],
                [c*a*math.cos(math.radians(beta)), c*b*math.cos(math.radians(alpha)), c**2]])

#print('The metric tensor for the given the given crystal system is',y)

#defining generalized inverse metric tensor
Y = np.linalg.inv(y)
#print('The generalized inverse metric tensor for this crystal system is',Y)

#defining slip system
#plane = (D1D21)= (801)
plane = np.array([8,0,1])
#slip_direction = [D3-D4 2]= [7,-7 2]
slip_dir = np.array([7,-7,2])
#load direction = [3 D5 D7] = [3 3 5]
loading_dir = np.array([3,3,5])


#calculations
p_l = np.dot(np.dot(plane,Y),loading_dir)
s_l = np.dot(np.dot(slip_dir,y),loading_dir)

#calculating norms
n1 = math.sqrt(np.dot(np.dot(plane, Y), plane.T)) * math.sqrt(np.dot(np.dot(loading_dir, Y), loading_dir.T))
n2 = math.sqrt(np.dot(np.dot(slip_dir,y),slip_dir.T))* math.sqrt(np.dot(np.dot(loading_dir,y),loading_dir.T))

phi = (p_l)/(n1)
lam = (s_l)/(n2)

print("The value of angle phi between the plane normal and the loading direction is:\n",phi)
print("The value of angle lamda between the slip direction and the loading direction is:\n",lam)

#calculating schmid factor

scmid = phi*lam
print("The schmid factor for the given system is:\n", scmid)

```
