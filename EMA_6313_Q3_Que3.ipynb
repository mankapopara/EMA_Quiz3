```python
#question-3 youngs's modulus caculations
#UFID = 8 0 7 7 3 6 5 3 = D1 D2 D3 D4 D5 D6 D7 D8

import numpy as np
import math

# Stiffness matrix values for cubic material (in Pa)
c11 = c22 = c33 = 280 * 10**9 #c11=2D1D2
c12 = c13 = c23 = c21 = c31 = c32 = 136 * 10**9 #c12=1D5D6
c44 = c55 = c66 = 87.7 * 10**9 #c44=8D3.D4

# Define the stiffness matrix for the cubic material
stiffness_Matrix = np.array([
    [c11, c12, c13, 0, 0, 0],
    [c12, c11, c13, 0, 0, 0], 
    [c13, c13, c11, 0, 0, 0],
    [0, 0, 0, c44, 0, 0],
    [0, 0, 0, 0, c44, 0],
    [0, 0, 0, 0, 0, c44]
])

print("[")

for row in stiffness_Matrix:
    print("  [" + " ".join(f"{val: .3e}" for val in row) + "],")

print("]")

# Calculate the compliance matrix by inverting the stiffness matrix
compliance_matrix = np.linalg.inv(stiffness_Matrix)

# Print the compliance matrix

print("The resultant compliance matrix for the given cubic material in (Pa^-1) is:")
print("[")
for row in compliance_matrix:
    print("  [" + " ".join(f"{val: .3e}" for val in row) + "],")
print("]")

# Define the direction vector for Young's modulus calculation in the [233] direction
n_vec = np.array([2, 3, 3])

# Calculate the normalized direction cosines for the [233] direction
norm_n_vec = np.linalg.norm(n_vec)
a11 = n_vec[0] / norm_n_vec  # Direction cosine with the x-axis
a12 = n_vec[1] / norm_n_vec  # Direction cosine with the y-axis
a13 = n_vec[2] / norm_n_vec  # Direction cosine with the z-axis

#
m = np.array([a11,a12,a13])

print("The direction cosines are:\n",m)

# Extract relevant values from the compliance matrix
s11 = compliance_matrix[0, 0]
s12 = compliance_matrix[0, 1]
s44 = compliance_matrix[3, 3]

# Calculate Young's modulus along the [233] direction using the compliance matrix and direction cosines
E = 1 / (s11 - 2 * (s11 - s12 - s44 / 2) * ((a11**2 * a12**2) + (a12**2 * a13**2) + (a13**2 * a11**2)))

# Print the final result
print("The Young's modulus along the [233] direction is:\n", E,"Pa")
```
