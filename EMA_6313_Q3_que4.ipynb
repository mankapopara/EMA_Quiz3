```python
#question-4 compliance matrix caculations
#UFID = 8 0 7 7 3 6 5 3 = D1 D2 D3 D4 D5 D6 D7 D8

import numpy as np
import math

# Stiffness matrix values for cubic material (in Pa)
c11 = c22 = c33 = 58*10**9
c44 = c55 = c66 = 20*10**9
c12 = c13 = c23 = c21 = c31 = c32 = 17* 10**9

# Define the stiffness matrix for the cubic material
stiffness_matrix = np.array([
    [c11, c12, c13, 0, 0, 0],
    [c12, c11, c13, 0, 0, 0],
    [c13, c13, c11, 0, 0, 0],
    [0, 0, 0, c44, 0, 0],
    [0, 0, 0, 0, c44, 0],
    [0, 0, 0, 0, 0, c44]
])

# Calculate the compliance matrix by inverting the stiffness matrix
compliance_matrix = np.linalg.inv(stiffness_matrix)

#formating 
print("The resultant compliance matrix for the given cubic material in (Pa^-1) is:")

print("[")

for row in compliance_matrix:
    print("  [" + " ".join(f"{val: .3e}" for val in row) + "],")

print("]")
```
